```python
X = df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors',
       'waterfront', 'view', 'condition',  'sqft_basement',
       'yr_built', 'yr_renovated']]
y = df['bathrooms']
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test,y_train,y_test = train_test_split(X,y,test_size = 0.4, random_state = 1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
lm.coef_
```

```python
lm.score(X,y)
```

```python
predictions = lm.predict(X_test)
```

```python
type(predictions)
```

```python
len(predictions
   )
```

```python
prediction_series = pd.Series(
    predictions,
    index=y_test.index,
    name='predicted_bathrooms'
)
```

```python
kd = pd.DataFrame({
    'actual_bathrooms': y_test,
    'predicted_bathrooms': prediction_series
})

kd.head()
```
