```python
import pandas as pd
import seaborn as sns
```

```python
dd = pd.read_csv("cancer_data.csv")
dd
```

```python
dd.info()
```

```python
dd.isna()
```

```python
dd['Unnamed: 32'].isnull().sum()
```

```python
dd.drop('Unnamed: 32',axis = 1, inplace = True)
```

```python
dd.drop('id',axis= 1,inplace =True)
dd
```

```python
dd.diagnosis = [1 if value == 'M' else 0 for value in dd.diagnosis]
dd
```

## normalize the values


```python
dd['diagnosis'] = dd['diagnosis'].astype('category',copy=False)
plot = dd['diagnosis'].value_counts().plot(kind='bar', title="Class distributions \n(0: Benign | 1: Malignant)")
fig = plot.get_figure()
dd.diagnosis
```

```python
X = dd.drop("diagnosis", axis= 1)
y = dd["diagnosis"]
```

```python
sns.heatmap(data=dd)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler_X = StandardScaler()
```

```python
X__ = scaler_X.fit_transform(X)
```

```python
X__
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X__,y,test_size = 0.3 , random_state=57)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
ll =LogisticRegression()
```

```python
ll.fit(X_train,y_train)
```

```python
predict = ll.predict(X_test)
```

```python
predict
```

```python
data_series = pd.Series(predict ,index=y_test.index, name="data_series")
```

```python
data = pd.DataFrame({"cancer_predicted": data_series,"Actual_patients":y_test})
data.head()
```

```python
from sklearn.metrics import accuracy_score
```

```python
Accuracy = accuracy_score(y_test,predict)
print((f'Accuracy: {Accuracy:}'))
```
