```python
import pandas as pd
import numpy as np
import matplotlib as pyplot
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
X = pd.DataFrame(iris['data'] , columns = ['sepal length in cm','sepal width','petal length','petal width'])
```

```python
y = iris['target']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state = 1)
```

```python
X_train
y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
t_ree = DecisionTreeClassifier()
```

```python
t_ree.fit(X_train,y_train)
```

```python
predict_1 = t_ree.predict(X_test)
type(predict_1)
type(y_test)
```

```python
tr_ee = pd.Series(predict_1,index=range(len(y_test)), name="data_series")
xx = pd.DataFrame({"predicted": tr_ee ,"actual " : y_test})
```

```python
xx
```

```python
from sklearn.metrics import confusion_matrix , classification_report
```

```python
cm = confusion_matrix(predict_1,y_test)
print(cm)
kn = classification_report(y_test,predict_1)
print(kn)
```
