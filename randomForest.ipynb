```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
tent = pd.read_csv("Travel.csv")
tent.shape
```

```python
tent['totalvistitng'] = tent['NumberOfPersonVisiting'] + tent [ 'NumberOfChildrenVisiting']
```

```python
tent.head()
```

```python
tent.head()
```

```python
tent.describe()
```

```python
tent.drop('CustomerID',inplace =True,axis =1)
```

```python
tent.isnull().sum()
```

```python
tent.select_dtypes(exclude = 'object').describe()

```

```python
tent.Age.fillna(tent.Age.mode()[0],inplace =True)
tent.DurationOfPitch.fillna(tent.DurationOfPitch.median(),inplace =True)
tent.NumberOfFollowups.fillna(tent.NumberOfFollowups.mode()[0],inplace =True)
tent.NumberOfTrips.fillna(tent.NumberOfTrips.median(),inplace =True)
tent.MonthlyIncome.fillna(tent.MonthlyIncome.median(),inplace=True)
tent.totalvistitng.fillna(tent.totalvistitng.median(),inplace=True)
tent.NumberOfChildrenVisiting.fillna(tent.NumberOfChildrenVisiting.median(),inplace=True)
tent.PreferredPropertyStar.fillna(tent.PreferredPropertyStar.mode()[0],inplace=True)
tent.TypeofContact.fillna(tent.TypeofContact.mode()[0],inplace=True)


```

```python
tent.isnull().sum()
```

```python
X = tent.drop('ProdTaken',axis = 1)
y = tent['ProdTaken']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 1)
```

```python
X_train
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
X_test
```

```python
non_numeric = X.select_dtypes(include = 'object').columns
numeric = X.select_dtypes(exclude = 'object').columns
onehotencoder= OneHotEncoder(drop='first')
standardscaler = StandardScaler()
preprocessor  = ColumnTransformer(
    [
        ('OneHotEncoder' , onehotencoder,non_numeric),
        ('StandardScaler', standardscaler,numeric)
    ]
)
```

```python
X_train=preprocessor.fit_transform(X_train)

```

```python
pd.DataFrame(X_train)
```

```python
X_test=preprocessor.transform(X_test)
```

```python
pd.DataFrame(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
model = {
    'random'  :RandomForestClassifier(),
    'gre'  :GradientBoostingClassifier(),
    'decision' :DecisionTreeClassifier(),
   
}
for i in range(len(list(model))):
    models = list(model.values())[i]
    models.fit(X_train,y_train)
    
```

```python
predict = models.predict(X_test)
```

```python
pf = pd.Series( predict,index = range(len(y_test)))
pg = pd.DataFrame({"predictions ":predict,"actual_values":y_test})
pg
```

```python
from sklearn.metrics import accuracy_score, classification_report,ConfusionMatrixDisplay
```

```python
accuracy_score(predict,y_test)
```

```python
kk =classification_report(predict,y_test)
print(kk)
```
